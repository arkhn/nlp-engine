```python
df = pd.read_csv(path + 'noms2008nat_txt.txt', sep='\t')
```

```python
df.columns
```

```python
df['frequency'] = df['_1931_1940'] + df['_1941_1950'] + df['_1951_1960'] + df['_1961_1970'] + df['_1971_1980'] + df['_1981_1990'] + df['_1991_2000']
```

```python
for col_name in list(df.columns):
    if col_name not in ['NOM', 'frequency']:
        del df[col_name]
```

```python
df.columns
```

```python
freq_name = df.groupby('NOM')['frequency'].sum()
freq_name.index.name = 'name'
freq_name.to_csv(path + 'name.csv', sep=',', header=['frequency'])
```
